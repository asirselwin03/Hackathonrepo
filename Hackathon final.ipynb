```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/DS/Downloads/diabetes.csv")
print(df.head())
print(df['Outcome'].value_counts())
```

```python
import matplotlib.pyplot as plt
```

```python
df[['Glucose', 'BMI', 'Age']].hist(bins=15, figsize=(10,6))
plt.tight_layout()
plt.show()
df['Outcome'].value_counts().plot(kind='bar')
plt.title("Diabetes Outcome Distribution")
plt.xticks([0,1], ['Non-Diabetic', 'Diabetic'], rotation=0)
plt.show()
```

```python
cols_with_zero = ['Glucose', 'BloodPressure', 'BMI', 'Insulin', 'SkinThickness']

for col in cols_with_zero:
    df[col] = df[col].replace(0, df[col].mean())
```
